**IDENTIFIERS**

Identifiers are special names used for identification.

Ex: Variable names, functions, methods, classes

**RULES OF IDENTIFIERS**

The only allowed characters are:
* A-Z
* a-z
* 0-9
* _ (Underscore)

Every identifier must start with an alphabet or an underscore, but it cannot start with a digit.

Identifiers are case sensitive:
 X = 10
 x = 10

There is no limit for an identifier's length, but it is recommended not to exceed 15 characters.

Keywords are not identifiers.

**KEYWORDS**

Keywords are also known as predefined words, reserved words, or important words.

In Python, there are 35 keywords available. All keywords must be in lowercase except for the following: `True`, `False`, and `None`.

### **DATA TYPES**

Data types represent the type of data stored in a variable. In Python, there are several primary built-in data types.

Python is a **dynamically typed language**, which means we are not required to declare data types manually; the interpreter identifies the type at runtime based on the assigned value.

* **int**: Integer values (e.g., 10, -5).
* **float**: Floating-point or decimal values (e.g., 10.5).
* **bool**: Boolean values representing `True` or `False`.
* **complex**: Complex numbers representing a real and imaginary part (e.g., `a+bj`).
* **str**: String values used for text. These can be enclosed in single, double, or triple quotes.

**Note:** Python also provides several specialized libraries that offer additional data structures.

**TYPE CONVERSIONS**

Converting one data type to another data type is known as type conversion (or type casting).

### **Type Conversion Examples**

Type conversion (also known as type casting) allows us to convert a value from one data type to another.

*   `int()`: Converts a value to an integer.
*   `float()`: Converts a value to a floating-point number.
*   `str()`: Converts a value to a string.
*   `bool()`: Converts a value to a boolean value.

````python
# The int() function can convert integers, floats, booleans, and numeric strings.
# It cannot convert float-like strings (e.g., "10.3"), alphabetic strings, or complex numbers.
# Errors are displayed below
````

````python
print(int(12.3))
print(int(True))
print(int(False))
print(int("10"))
````
