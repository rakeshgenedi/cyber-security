**IDENTIFIERS**

Identifiers are unique names used for identification.

Ex: Variable names, functions, methods, classes

**RULES FOR IDENTIFIERS**

The only allowed characters are:
* A-Z
* a-z
* 0-9
* _ (Underscore)

Every identifier must start with an alphabet character or an underscore; it cannot start with a digit.

Identifiers are case-sensitive:
 X = 10
 x = 10

There is no limit for an identifier's length, but it is recommended not to exceed 15 characters.

Keywords cannot be used as identifiers.

### **DATA TYPES**

Data types represent the type of data stored in a variable. In Python, there are several primary built-in data types.

Python is a **dynamically typed language**, which means we are not required to declare data types manually; the interpreter identifies the type at runtime based on the assigned value.

* **int**: Integer values (e.g., 10, -5).
* **float**: Floating-point or decimal values (e.g., 10.5).
* **bool**: Boolean values representing `True` or `False`.
* **complex**: Complex numbers representing a real and imaginary part (e.g., `a+bj`).
* **str**: String values used for text. These can be enclosed in single, double, or triple quotes.

**Note:** Python also provides several specialized libraries that offer additional data structures.

**Operators**

An operator is a symbol which performs operations on one or more operands.

In Python, there are different types of operators:

1. **Arithmetic Operators**:
 `[+, -, *, %, /, //, **]`
2. **Assignment Operators**:
 `[=, +=, -=, *=, /=, **=, //=]`
3. **Relational Operators**:
 `[>, <, ==, >=, <=, !=]`
4. **Special Operators**:
 `[in, not in, is, is not]`
5. **Bitwise Operators**:
 `[&, |, ^, ~, <<, >>]`
6. **Logical Operators**:
 `[and, or, not]`
7. **Conditional (Ternary) Operators**

```python
# Arithmetic operators
# Priority order
# **
# *, /, //, %
# +, -

print(12 + 4)
print(12 - 4)
print(12 / 4)
print(12 // 4)
print(12 & 4)
print(12 // 4)
```

```python
# expression example

print(5+6*2//7+9%4*2-1)
print(7//3*2+8%3+7//5*3)
print(5//8*3+4%9+3//7*3+2%7)
```

```python
# Assignment operator

a = b
a = b + c
a = a + b
a += b
x = x * y
x *= y
n = n / 10
n /= 10
p = p - q
p -= q
```

```python
#  Relational operator

a = 10
b = 20
c = 10

print(f"a = {a}, b = {b}, c = {c}")
print(f"a > b: {a > b}")
print(f"a < b: {a < b}")
print(f"a == c: {a == c}")
print(f"a >= c: {a >= c}")
print(f"b <= a: {b <= a}")
print(f"a != b: {a != b}")
```

**Special operators**

In Python, there are 2 types of special operators:
1. Membership operators: `in`, `not in`

2. Identity operators: `is`, `is not`



```python
# Membership operators

s ="welcome to cat computer point"
print('e' in s)
print('me' in s)
print('cat' in s)
print('z' in s)
print('z' not in s)
print('we' not in s)
```

```python
# identity operator

x =10
y= 10
z =20
print(id(x))
print(id(y))
print(id(z))
print(x is y)
print(x is not y)
print(x is z)
print(x is not z)
```
